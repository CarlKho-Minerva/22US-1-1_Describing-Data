```python
def carlmean(numList):
    # initialize total to 0 representing the variable that will hold the sum of all the numbers in the list
    total = 0
    for num in numList:             # Iterate through the values in list
        # Repeatedly add the values in the list to the total one by one until there are no more values in the list
        total += num
    # Return the average by dividing the total values inside the list and the total quantity (length) of the list.
    return total/len(numList)
    # Did not store the average in a variable because it was only used once.
```

```python
def carlmedian(numList):

    # a method to sort the list in ascending order
    numList.sort()
    # variable to get the length of the list to reduce usage of len(numList) throughout the code
    numList_len = len(numList)

    # IF odd or IF the total quantity of a number list (len()) is not divisible by 2 (calculated with the number
    if numList_len % 2 != 0:
        # list modulo (%) 2 NOT zero)

        # "Index" is defined as the position of a VALUE inside a list. Here, floor division was used to convert the float output into an integer in a cleaner manner (instead of using int())
        middle_index = (numList_len - 1) // 2
        # I also used the -1 to account for the fact that the index starts at 0

        # return the value of the determined middle index of the list
        return numList[middle_index]

    # ELSE (if the total quantity of a number list (len()) is divisible by 2 (calculated with the number list modulo (%) 2 zero))
    elif numList_len % 2 == 0:

        # For this case, we need to find the average of the two values of the two middle indexes of numList
        # The first middle index is the total quantity of the list divided (floor) by 2
        middle_index_1 = numList_len // 2
        # The second middle index is the first middle index minus 1
        middle_index_2 = middle_index_1 - 1

        # return the integer average of the two middle indexes of the list
        return (numList[middle_index_1] + numList[middle_index_2]) // 2
```

```python
def carlmode(numList):

    # initialize the max_count variable to a tuple of 0s to represent the variable that will hold the max count of a number in the list
    # we use a tuple because it is immutable or the value cannot be changed, and we do not want to change the value of the max_count variable because it is used to compare the count of the other numbers in the list to determine the mode.
    max_count = (0, 0)

    # iterate through the values in the list
    for num in numList:
        # variable to hold the number of times a value appears in the list
        # we use the count() method to count the number of times a value appears in the list
        # count works by counting the number of times a value appears in the list and returning the count
        instancesValue = numList.count(num)
        # IF the number of times a value appears in the list is greater than the max count of a number in the list
        if instancesValue > max_count[1]:
            # set the max_count variable to the value and the number of times it appears in the list
            max_count = (num, instancesValue)
            # we do this because we want to compare the count of the other numbers in the list to determine the mode so that we can return the mode at the end of the function.

    # return the value of the max count of a number in the list
    return max_count[0]
```

```python
def carlrange(numList):
    # a method to sort the list in ascending order
    # sort works by comparing the first two values in the list and swapping them if the first value is greater than the second value (ascending order).
    numList.sort()
    # return the range of the list
    # the range is the difference between the first and last values in the list
    return numList[-1] - numList[0]
```

```python
def carlstandardDeviation(numList):

    # did not use mean in code cell 3 because calculating int and function = error
    # the sum function works by adding all the values in a list
    mean = sum(numList) / len(numList)
    # variable to hold the variance of the list. SUM used here...

    # the variance works by subtracting the mean from each value in the list, squaring the result, and then adding all the values together and dividing by the total quantity of values in the list.
    variance = sum([((x - mean) ** 2) for x in numList]) / len(numList)

    # return the standard deviation of the list
    # the standard deviation works by taking the square root of the variance, which is why the variance is calculated first. We simply multiply the variance by 0.5 to get the square root of the variance.
    return variance ** 0.5
```

(2.1) What is the probability of obtaining three entries that satisfy A and four that do not, P(three A and four A^c)? Hint: can you use the binomial distribution? Show your work and explain your steps sufficiently. 

------------------ CARL'S ANSWER ------------------
Yes, we can use the binomial distribution because we are looking for the probability of observing x successes in N trials.



The factorial function with a loop works by multiplying the result by i each time through the loop. The loop starts at 1 and ends at n+1 because the range function does not include the last number in the range. The result is initialized to 1 because the factorial of 0 is 1.



The binomial function works by calculating the combination of n choose k, which is the number of ways to choose k items from a set of n items. The combination is calculated by dividing the factorial of n by the product of the factorial of k and the factorial of n-k. The probability of success to the power of k successes is calculated by raising the probability of success to the power of k. The probability of failure to the power of n-k failures is calculated by raising the probability of failure to the power of n-k. The combination, probability of success, and probability of failure are then multiplied together to get the binomial distribution.

```python
# # RECALL Probability calculations for the following events for future calculations:
# A = grade_10andBelow_prob                                         # 0.51  (A)                 # PROBABILITY OF SUCCESS
# notA = grade_10andBelow_prob_inverse                              # 0.49  (A^c)               # PROBABILITY OF FAILURE
# B = strength_training_3orMore_prob                                # 0.50  (B)
# notB = strength_training_3orMore_prob_inverse                     # 0.50  (B^c)
# AandB = grade_10andBelow_strength_training_3orMore_prob           # 0.255  (A & B)

# factorial function with loop
def factorial(n):
      result = 1                    # initialize result
      # loop from 1 to n and multiply result by i each time through the loop.  # SAMPLE range(1, 5) = [1, 2, 3, 4] to account for 0 index in python lists
      for i in range(1, n+1):
            result *= i             # result = result * i
      return result

# binomial distribution function with made factorial function. Follows the analytical formula for binomial distribution.
def binomial(n, k, p):
      combination = factorial(n) / (factorial(k) * factorial(n-k))
      probSuccess = (p**k) # probability of success to the power of k successes (k successes)
      probFailure = ((1-p)**(n-k)) # probability of failure to the power of n-k failures (n-k failures)
      return combination * (probSuccess * probFailure)


# calculate probability of choosing three A and four A^c with a binomial distribution function
n = 7       # number of trials
p = A       # probability of success (A) = 0.51
k = 3       # number of successes being asked = 3
print("Probability of choosing three A and four A^c: ",
      binomial(n, k, p) * 100, "%")

# check the probability of choosing three A and four A^c with the binomial distribution function from scipy
# import scipy binomial distribution
from scipy.stats import binom

#  binom.pmf works by taking the number of successes, number of trials, and probability of success as arguments and returns the probability of choosing k successes in n trials with a probability of success p.
threeA_fourNOTA_SCIPY = binom.pmf(k, n, p)

print("Probability of choosing three A and four A^c: ",
      threeA_fourNOTA_SCIPY * 100, "%")
```
